```python
import matplotlib.pyplot as plt

attrition_counts = df['Attrition'].value_counts()

colors = ['#ff6666', '#ffcc99']

plt.pie(attrition_counts, explode=(0.1, 0), colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)

plt.title('Attrition Rate')
plt.legend(labels=attrition_counts.index, loc='upper right')

plt.show()
```



The majority of employees (around 83%) do not have Attrition. This information can be useful in understanding the overall of the employee population and identifying potential factors that contribute to job satisfaction and retention

```python
grouped_data2 = df.groupby(['Department', 'Attrition'])['Attrition'].count()

df_group2=grouped_data2.unstack()
df_group2
```

```python
df_group2.plot(kind='bar', stacked=True, color=['#ff6666', '#ffcc99'])

plt.title('Attrition Rate by Department')
plt.xlabel('Department')
plt.ylabel('Count')

plt.show()
```


- Research & Development department has the lowest attrition rate.
- Attrition rate is higher for employees in Sales and Marketing departments compared to other departments.
-Sales department has the highest attrition rate.

```python
grouped_data = df.groupby(['MaritalStatus', 'Gender'])['Attrition'].count()

df_group=grouped_data.unstack()
df_group
```

```python
df_group.plot(kind='bar', stacked=True, color=['#ff6666', '#ffcc99'])

plt.title('Attrition Rate by Marital Status and Gender')
plt.xlabel('Marital Status and Gender')
plt.ylabel('Count')

plt.show()
```

- Male employees with MaritalStatus as 'Single' have the highest attrition rate. This could be due to several reasons, such as:
    - Single employees may have more opportunities for career advancement outside of the current company.
    - Single employees may have more financial flexibility to leave their current job if they are not satisfied.
    - Single employees may have more personal responsibilities, such as caring for children or elderly parents, that make it difficult to stay in a job that is not meeting their needs.
- Female employees with MaritalStatus as 'Married' have the lowest attrition rate. This could be due to several reasons, such as:
    - Married employees may have more financial stability and less need to leave their current job for financial reasons.
    - Married employees may have more support from their spouse and family, which can help them to cope with the stresses of work and make them less likely to leave their job.
    - Married employees may have more opportunities for career advancement within the current company.
- Attrition rate is higher for single employees compared to married employees. This is likely due to a combination of the factors mentioned above.

```python
palette = sns.color_palette("Set2", 10)

grouped_data = df.groupby(['JobRole', 'MonthlyIncome'])['MonthlyIncome'].count()

sns.boxplot(x='JobRole', y='MonthlyIncome', data=df, order=['Healthcare Representative', 'Research Scientist', 'Laboratory Technician', 'Manufacturing Director', 'Healthcare Representative', 'Manager', 'Sales Executive', 'Research Director', 'Sales Representative', 'Human Resources'], palette=palette)

plt.title('Monthly Income by Job Role')
plt.xlabel('Job Role')
plt.ylabel('Monthly Income')

plt.xticks(rotation=90)

plt.show()
```


- Research Director and Healthcare Representative have the highest median
monthly income.
- Laboratory Technician and Research Scientist have the lowest median monthly income.
- There is a wider range of monthly incomes for Sales Executive and Sales Representative compared to other job roles.
- There are a few outliers with very high monthly incomes for Research Director,Human Resource,Healthcare reprentative and Sales Executive.

```python
column=['Age', 'Attrition', 'BusinessTravel', 'DailyRate', 'Department',
       'DistanceFromHome', 'Education', 'EducationField', 'EmployeeCount',
       'EmployeeNumber', 'EnvironmentSatisfaction', 'Gender', 'HourlyRate',
       'JobInvolvement', 'JobLevel', 'JobRole', 'JobSatisfaction',
       'MaritalStatus', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked',
       'Over18', 'OverTime', 'PercentSalaryHike', 'PerformanceRating',
       'RelationshipSatisfaction', 'StandardHours', 'StockOptionLevel',
       'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance',
       'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion',
       'YearsWithCurrManager']
correlation_matrix = df[column].corr()

# Plotting the heatmap
plt.figure(figsize=(20, 15))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap', fontsize=16)
plt.show()
```

- There is a positive correlation between Age and TotalWorkingYears.
- There is a positive correlation between JobLevel and TotalWorkingYears.
- There is a positive correlation between TotalWorkingYears and MonthlyIncome.
- There is a negative correlation between DistanceFromHome and JobSatisfaction.
- There is a positive correlation between YearsInCurrentRole and YearsWithCurrManager.
